Income Buckets are usually divided into \\$25k/yr increments, our customer data is in USD/month not year. converting   \\$25k/yr into monthly income is approx \\$2k. With a range of \\$0 to \\$35k/month, it would be 17 categories. 17 can be reduced to 12 or 13 because the data has a distinct line hovering about \\$25k/month, say \\$25k+/month.

```python
income = customer_df[['Income (USD/Month)']]


filters = []
for a in range(2,6,1):
    b = a*1000
    bucket= dict()
    for _, i in enumerate(range(b,37000,b)):
        if _ == 0:
            x = income[income["Income (USD/Month)"]<=i].count()["Income (USD/Month)"]
            bucket[f"<{i//1000}k"] =  x
        else:
            x = income[(income["Income (USD/Month)"]<i) & (income["Income (USD/Month)"]>=i-b)].count()["Income (USD/Month)"]
            bucket[f"{(i-b)//1000}k - {(i/1000)-0.1}k"] = x
            
    filters.append(pd.Series(bucket))
filters[0].plot.bar()
```

```python
filters[1].plot.bar()
```

```python
filters[2].plot.bar()
```

```python
filters[3].plot.bar()
```

filters[1] the \\$3k increments is the most usable bucketing. buckets will range from <$3k to >=$33k with 11 bucket in total.

```python
def income_bucket(value):
    bucket_value = value//3000
    if bucket_value ==0:
        bucket = f"Under ${(bucket_value+1)*3}k"
    elif bucket_value >= 11:
        bucket = f"${(bucket_value)*3}k+"
    else:
        bucket = f"${bucket_value*3}k to ${(bucket_value+1)*3}k"
    return bucket
```

```python
customer_df["Bucket_Value"] = customer_df['Income (USD/Month)'].apply(lambda x: x//3000 )
customer_df["Income Bucket"] = customer_df["Income (USD/Month)"].apply(lambda x: income_bucket(x))
customer_df
```

```python
#total Rides per income buket
customer_df.groupby("Bucket_Value").sum()["Total Rides"].plot.bar()
plt.ylabel("Number of People")
plt.title("Count of People per Income Bucket")
```

```python
# count of people per bracket
customer_df.groupby("Bucket_Value").count()["Total Rides"].plot.bar()
plt.ylabel("Ride Count")
plt.title("Count of Cab Rides per Income Bucket")
```

The data distribution of "Count of People per Income Bucket" is the same as the "Ride Count per Income Bucket".
